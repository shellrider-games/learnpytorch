Get my GPU

```python
import torch
```

```python
device = 'cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu'
```

```python
tensor = torch.rand(1000,1000,1000)
tensor.device
```

```python
#Move tensor to gpu
tensor_on_gpu = tensor.to(device)
```

```python
tensor_square_on_gpu = tensor_on_gpu * tensor_on_gpu
```

`tensor_square_on_gpu.numpy()`
code errors cause tensor is stored on gpu

```python
#Get it back onto cpu
tensor_back = tensor_square_on_gpu.cpu()
tensor_back.numpy()
```
