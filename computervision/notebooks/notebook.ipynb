## 1. Getting a dataset

Fashion MNIST

```python
train_data = datasets.FashionMNIST(
    root="datasets/fashion_mnist",
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=None
)
```

```python
test_data = datasets.FashionMNIST(
    root="datasets/fashion_mnist",
    train=False,
    download=True,
    transform=ToTensor(),
    target_transform=None
)
```
